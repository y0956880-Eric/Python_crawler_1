```python
# BMI = 體重(公斤) / (身高(公尺))^2

weight = 70      # 體重，單位：公斤
height = 1.75    # 身高，單位：公尺

bmi = weight / (height ** 2)
bmi = round(bmi, 2)
print("BMI:", bmi)
```
