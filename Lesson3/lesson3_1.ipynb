```python
# 字串和變數的整合
apple = 5
orange = 9

"蘋果=" + str(apple) + "個,橘子=" + str(orange) + "個"
```

```python
# str template
# 字串差補變數
# 字串和變數整合,傳出1個全新的字串
print(f"蘋果={apple}個,橘子={orange}個")
```
